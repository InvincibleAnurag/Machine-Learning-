# Linear Regression implementation model for House-Price Prediction System

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("homeprices.csv")
#df.head(2)
#df.tail()
#df.head()
#df.tail(2)
#df.describe()
#df.info()
#df.shape
df
```

```python
%matplotlib inline
plt.xlabel("Area(sq.ft)")
plt.ylabel('Price($)')
plt.scatter(df['area'],df.price,color='red',marker='*')
df['area']
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
new_data=np.array(3300).reshape(-1,1)
new_data
y_pred=reg.predict(new_data)
y_pred
new_data1=np.array(3700).reshape(-1,1)
new_data1
y_pred1=reg.predict(new_data1)
y_pred1
```

```python
print("Cofficient{}:",reg.coef_)
```

```python
print("Intercept{}:",reg.intercept_)
```

```python
%matplotlib inline
plt.xlabel("Area(sq.ft)")
plt.ylabel('Price($)')
plt.scatter(df.area,df.price,color='red',marker='*')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
```
